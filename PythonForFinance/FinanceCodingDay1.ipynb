```python
#All of the below has been taken from the book Python For Finance, O'Reily
import numexpr as ne
import numpy as np
import pandas as pd
from statsmodels import regression
import statsmodels.api as sm
import matplotlib.pyplot as plt
import math
import pandas.io.data as web
%matplotlib inline
```

I will be using Black Scholes Merton Model to calculate implied volatities. The inputs are [S,K,t,r,sigma]:
1. S= price of the underlying at t
2. K= Strike price of the option
3. t= time to maturity
4. r= riskless short rate
5. sigma= constant volatility

```python
sp500 = web.DataReader("^GSPC",data_source="yahoo",start="1/1/2000",end="4/14/2014")
```

```python
sp500.info()
```

```python
sp500['Close'].plot(grid=True, figsize=(8, 5))
```

```python
sp500['42'] = np.round(pd.rolling_mean(sp500['Close'],42),2)
sp500['252'] = np.round(pd.rolling_mean(sp500['Close'],252),2)
```

```python
sp500['42'].name = 'pr'
sp500[['42','252','Close']].plot()
```

```python
sp500['42-252'] = sp500['42']-sp500['252']
sp500['42-252'].tail()
```

```python
SD = 50;
sp500['Regime'] = np.where(sp500['42-252']>SD, 1, 0)
print sp500['Regime']
sp500['Regime'] = np.where(sp500['42-252']<-SD, -1, sp500['Regime'])
```

```python
sp500['Regime'].value_counts()
```

```python
sp500['Regime'].plot()
```

```python
sp500["Market"] = np.log(sp500['Close']/sp500['Close'].shift(1))
#I need the market first 
sp500['Strategy'] = sp500['Regime'].shift(1) * sp500['Market']
```

```python
sp500[['Market','Strategy']].cumsum().apply(np.exp).plot(grid=True, figsize=(5, 8))
```
